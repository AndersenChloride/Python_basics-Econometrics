## Авторегрессионная модель
Авторегрессионная модель (AR-модель) – это статистическая модель, которая используется для анализа временных рядов. Она основывается на предположении, что текущее значение временного ряда зависит от предыдущих значений этого же ряда.

AR-модель может быть представлена в виде уравнения, где текущее значение временного ряда является функцией от предыдущих значений этого же ряда и случайной ошибки. Коэффициенты уравнения определяются методом наименьших квадратов, который позволяет найти такие значения коэффициентов, чтобы минимизировать сумму квадратов ошибок.

$x_{t} = c + α_{1}x_{t-1}+α_{2}x_{t-2}+...+α_{p}x_{t-p}+ε_{i}$

AR-модель может быть использована для прогнозирования будущих значений временного ряда, а также для анализа влияния различных факторов на этот ряд.



## Метод скользящего среднего
Метод скользящего среднего (MA) - это метод анализа временных рядов, который используется для сглаживания данных и выявления трендов. Он основывается на вычислении среднего значения временного ряда за определенный период времени.

$SMA_{t}=\frac{1}{n}\sum_{i=0}^{n-1}x_{t-i}$

Для применения метода скользящего среднего необходимо выбрать длину окна, которое представляет собой количество последовательных значений временного ряда, на которые будет распространяться сглаживание. Затем для каждого окна вычисляется среднее значение, которое заменяет исходное значение временного ряда в центре окна.

Применение метода скользящего среднего позволяет устранить случайные колебания в данных и выделить более четкие тренды. Однако этот метод может также сгладить значимые изменения в данных, такие как пики или провалы.


## Модель ARMA

При анализе временных рядов модель ARMA (модель авторегрессионного-скользящего среднего) обеспечивает описание стационарного случайного процесса во времени в теоминах двух вышерассмотренных полиномов: полином авторегрессии, включающий регрессию переменной к её собственным запаздывающим значениям, и полином скользящего среднего, включающий линейную комбинацию членов ошибки.

$ X_{t}=c + \sum_{i=1}^{p}φ_{i}x_{t-i} + \sum_{j=1}^{q} θ_{j} ε_{t-j}$  

$φ_{i}, θ_{j}$ - коэффициенты при аргументах, которые могут представлять из себя весы для корректирования модели, если требуется сделать акцент на определенных значениях.

с - константа, представляющая собой постоянный член, который не зависит от других переменных в модели и добавляется к каждому значению зависимой переменной. Она может помочь учесть факторы, которые не объясняются другими переменными в модели, но могут влиять на зависимую переменную. Например, если мы моделируем продажи товаров, константа может отражать общий уровень спроса на рынке, который не зависит от других факторов, таких как цена или реклама.

ARMA модель строится на стационарном временном ряде. Данное условие редко выполняется в реальности, поэтому существуют способы приведения нестационарного ряда к стационарному. В данной работе будет использован метод дифференцирования.

Одной из основных проблем построения модели является подбор параметров p,q. Значения этих парамеров характеризуют глубину модели. Слишком высокое значение создаст лишние шумы, в то время как слишком низкое не позволит создать точную модель. Для решения этой проблемы существуют различные методы подбора оптимальных параметров.

##Проверка на выбросы. Тест Граббса

Тест Граббса является статистическим инструментом для выявления выбросов в наборе данных. Он был предложен американским статистиком Фрэнком Граббсом в 1969 году и является одним из наиболее распространенных тестов на выбросы.

Суть теста Граббса заключается в сравнении наибольшего или наименьшего значения в выборке с остальными значениями. Если это значение лежит дальше от среднего значения выборки, чем ожидалось для нормального распределения, то оно считается выбросом.

$U_{расч} = \frac{|X_{i}-X_{ср}|}{S}$

Также из библиотеки scipy загрузим функцию для рассчета t-критерия.

```python
from scipy.stats import t

def grubbs_test(data, alpha):
    n = len(data)
    t_critical = t.ppf(1 - alpha / (2 * n), n - 2)
    mean = np.mean(data)
    std = np.std(data, ddof=1)
    max_deviation = np.max(np.abs(data - mean))
    G = max_deviation / std
    is_outlier = G > ((n - 1) / np.sqrt(n)) * np.sqrt(t_critical ** 2 / (n - 2 + t_critical ** 2))
    return is_outlier

go = data['Price']
go = go.tolist()

outliers_math = grubbs_test(go,0.05)

print("Выбросы:", outliers_math)
```

Проверка на выбросы прошла успешно при  $\alpha$=0,05. Выбросы отсутствуют

Стационарность - это свойство случайной величины или процесса сохранять свои статистические характеристики (например, математическое ожидание или дисперсию) во времени.

###Тест Дики-Фуллера
Для того, чтобы точно убедиться, что ряд стационарным не является применим статистику Дики-Фуллера.  
Тест Дики-Фуллера был первым статистическим тестом, разработанным для проверки нулевой гипотезы о том, что единичный корень присутствует в авторегрессионной модели заданного временного ряда и что процесс, таким образом, не является стационарным.

$y_{i}=δy_{i-1}+\varepsilon_{i}$  

где $δ$ — параметр модели, $\varepsilon_{i}$ - белый шум (данные, влияние, которых мы не можем учесть)

Такой процесс является стационарным при условии $|δ|\lt 1$.

```python
from statsmodels.tsa.stattools import adfuller

result1 = adfuller(data['Price'])
print('ADF Statistic: %f' % result1[0])
print('p-value: %f' % result1[1])
print('Critical Values:')
for key, value in result1[4].items():
    print('\t%s: %.3f' % (key, value))
```

По результатам видно, что p-значение сильно больше $\alpha$. Подсчитанное значение больше критических при различных $\alpha$. Таким образом, гипотеза о стационарности ряда отвергается при всех рассмотренных $\alpha$.

###Дифференцирование
Дифференцирование ряда – это математическая операция, которая позволяет перейти от исходного ряда к его производной функции. В контексте достижения стационарности, дифференцирование ряда может использоваться для устранения тренда или сезонности, что позволяет упростить анализ временных рядов.  

Если исходный ряд содержит единичный корень, то его первая разность (т.е. разность между соседними значениями ряда) может не содержать единичных корней и стать стационарным. Если первая разность ряда также содержит единичный корень, то можно продолжать дифференцирование до тех пор, пока все единичные корни не будут устранены и ряд не станет стационарным.

Применим этот метод к нашей выборке

```python
def difference(data, interval=1):
    diff = []
    for i in range(interval, len(data)):
        value = data[i] - data[i - interval]
        diff.append(value)
    return np.array(diff)

points = data['Price']
points = difference(points)

plt.plot(data['Data'][:547], points)
plt.xlabel('Date', fontsize=15)
plt.ylabel('Value', fontsize=15);
plt.show()
```

Проведем тест Дики-Фуллера над новой выборкой.

```python
result2 = adfuller(points)
print('ADF Statistic: %f' % result2[0])
print(f"p-value: {result2[1]}")
print('Critical Values:')
for key, value in result2[4].items():
    print('\t%s: %.3f' % (key, value))
```

В этот раз p-значение меньше $\alpha$. Новое значение статистики меньше критических. Принимаем гипотезу о стационарности нового ряда.

##Построение ARMA-модели
Для начала разделим выборку на тренировочную и тестовую часть.

```python
train_size = int(len(points) * 0.8)
train, test = points[:train_size], points[train_size:]
```

Подберем параметры навскидку:  
p=6  
q=6

Построим модель ARMA с глубиной равной 6 на тренировочной выборке и предскажем будущие значения с помощью тестовой.

```python
p = 6
d = 0
q = 6

model1 = ARIMA(train, order=(p,d,q)).fit()
predictions = model1.forecast(steps=len(test))
```

```python
mse1 = mean_squared_error(test, predictions)
mse2 = mean_absolute_error(test, predictions)
print('MSE: %.3f' % mse1)
print('MAE: %.3f' % mse2)
```

Итак, мы построили модель ARMA, однако выбранные навскидку параметры могут быть не лучшими для построения модели по нашим данным, то есть, возможно, есть параметры, которые бы смогли дать нам лучшую точность. Для выбора таких параметров рассмотрим различные методы их определения.

##Критерий Акаике
В процессе поисков различных методов было решено обратится к критерию Акаике, предложенном Хиротугу Акаике в 1971 году. Критерий основан на оценке потери информации при уменьшении числа параметров модели. Критерий позволяет найти компромисс между сложностью модели (числом параметров) и ее точностью.

$AIC=2k - 2ln(L) $,

где k — число параметров модели, L — максимизированное значение функции правдоподобия модели. Лучшей признается та модель, для которой значение AIC минимально.

Данный критерий характеризует собой объем потерянных данных при приближении рассматриваемой модели. Таким образом, имея большое количество моделей мы можем оценить каждую и выбрать ту, которой соответствует наименьшее значение критерия, то есть ту, что имеет наименьшую потерю данных.
Будем предполагать, что ошибки модели нормально и независимо распределены. Пусть n – число наблюдений (число точек известных для таблично заданной функции), а RSS – остаточная сумма квадратов. Предположим, что дисперсия наших ошибок одинакова. Тогда формула приобретает следующий вид:  

$ AIC=2k + n[ln(((2πRSS+1)/n)+1)] $  

Так как наши модели будут содержать одинаковое количество точек, упростим это выражение

$AIC=2k + n[ln(RSS)]$  

где k = (p + q + 1), n — количество точек, RSS - остаточная сумма квадратов.

```python
import warnings
warnings.filterwarnings("ignore")
p = range(0, 6)
q = range(0, 6)

models = []
for pi in p:
     for qi in q:
          model = ARIMA(train, order=(pi, 0, qi))
          models.append((model.fit().aic, pi, 0, qi))

best_model = min(models)
print('Best model: AIC = {}, p = {}, d = {}, q = {}'.format(*best_model))
```

Критерий был использован для подбора лучших параметров в диапазоне от 0 до 4.
Лучшими параметрами по критерию оказались  
p=2  
q=1

Как видно, нам удалось улучшить точность модели. Использование критерия было оправдано. Построим график, чтобы визуально оценить модель авторегрессии скользящего среднего.

##Оценка модели с помощью метода one-step кросс-валидации
One-step кросс-валидация (или "one-ahead out-of-sample validation") - это метод оценки качества работы модели временного ряда, который заключается в последовательном обновлении модели на каждом шаге времени и тестировании ее на следующем шаге.

Суть метода заключается в том, что на каждом шаге времени используется накопленная информация до этого момента для обучения модели, после чего эта модель применяется для прогнозирования значения следующего шага. Это повторяется для всех доступных шагов времени, тестирование проводится на данных, которые модель ранее не видела.

Такой подход позволяет оценить точность прогнозов на практике, когда данные поступают постепенно и неизвестны заранее. Также one-step кросс-валидация помогает проверить, насколько хорошо модель обобщает данные, так как она тестируется на новых данных, не участвующих в обучении.

Однако, следует учитывать, что использование one-step кросс-валидации может быть достаточно ресурсоемким, так как требует многократной перетренировки модели.


Для реализации кросс-валидации для ARMA модели в Python можно использовать функцию TimeSeriesSplit библиотеки sklearn.model_selection.

```python
time_series = points
horizon = 10

p1 = 4
q1 = 5

p2 = 6
q2 = 6

n_splits = 10


train_size = int(len(time_series) * 0.8)

errors1 = []
errors2 = []


for i in range(n_splits):
    train_data = time_series[:train_size]
    test_data = time_series[train_size:train_size+horizon]

    model_fit1 = ARIMA(train_data, order=(p1,0,q1)).fit()
    model_fit2 = ARIMA(train_data, order=(p2,0,q2)).fit()


    predictions1 = model_fit1.predict(start=len(train_data), end=len(train_data)+len(test_data)-1)
    predictions2 = model_fit2.predict(start=len(train_data), end=len(train_data)+len(test_data)-1)

    error1 = mean_squared_error(test_data, predictions1)
    errors1.append(error1)

    error2 = mean_squared_error(test_data, predictions2)
    errors2.append(error2)

    train_size += horizon

mean_error1 = np.mean(errors1)
print('Mean squared error:', mean_error1)

mean_error2 = np.mean(errors2)
print('Mean squared error:', mean_error2)
```

По данному тесту мы можем увидеть, что ошибка у модели с параметрами p=4, q=5 меньше, чем у той, что была подобрана навскидку. Для того, чтобы точно убедиться, какая модель лучше, проведем анализ остатков.

##Оценка остатков модели. Автокорреляция остатков
Остатки модели - это разница между фактическим значением целевой переменной и ее прогнозируемым (модельным) значением. Если модель хорошо соответствует данным, то остатки должны быть случайными, распределенными нормально и без корреляции между собой.

Для проверки наличия корреляции между остатками модели и временем можно использовать графики автокорреляции (ACF) и частичной автокорреляции (PACF). График ACF показывает корреляцию между остатками на разных лагах, тогда как график PACF отображает только прямую зависимость между остатками на разных лагах. Анализ этих графиков может помочь выявить наличие автокорреляции в остатках модели.


$ \mathrm{ACF}(k) = \frac{1}{n} \sum_{i=1}^{n-k} \left[\left(y(i) - \overline{y}\right) \cdot \left(y(i+k) - \overline{y}\right)\right]   $

$ \mathrm{PACF}(k) = \begin{cases} \mathrm{corr}(y(1), y(k)), & k=1 \\ \frac{\mathrm{corr}\left(y(k) - \hat{y}_k, y(k-j) - \hat{y}_{k-j} \right)}{\mathrm{corr}(y(k-j), y(k-j))}, & k>1, j=1,\ldots,k-1 \end{cases}  $  

Если на графиках наблюдаются значимые значения корреляции, это может указывать на наличие структуры в остатках, которую нужно учитывать при построении модели. Например, если значения ACF значительны на первых нескольких лагах, это может указывать на наличие авторегрессионной зависимости в данных. Если же значения PACF значительны только на первом лаге, это может указывать на присутствие пропущенных факторов в модели.

Одним из способов получить остатки модели является использование метода наименьших квадратов (OLS). Для этого нужно запустить модель регрессии, затем получить прогнозы для целевой переменной на основе наблюдаемых значений признаков и вычислить разницу между фактическими значениями целевой переменной и ее прогнозируемыми значениями.

$ \begin{equation}
S = \sum_{i=1}^{n} e_i^2 = \sum_{i=1}^{n} (y_i - X_i\beta)^2
\end{equation}  $



## Тест Бройша-Пагана
Тест Бройша-Пагана является одним из наиболее широко используемых тестов на гетероскедастичность в эконометрике. Этот тест был разработан в 1982 году американскими экономистами Дэвидом Бройшем и Ричардом Паганом и представляет собой модификацию теста Уайта.

Гетероскедастичность – это явление, при котором дисперсии случайных ошибок регрессии неодинаковы для всех значений объясняющих переменных. Это значит, что некоторые значения объясняющих переменных приводят к большей дисперсии ошибок, чем другие значения. Гетероскедастичность может привести к неправильной интерпретации результатов оценки модели, а также к снижению точности оценок параметров.

Тест Бройша-Пагана заключается в следующем: сначала оценивается регрессия, затем вычисляются квадраты остатков этой регрессии и оценивается регрессия этих квадратов на все объясняющие переменные и константу. Затем вычисляется статистика теста, которая имеет распределение хи-квадрат с числом степеней свободы, равным количеству объясняющих переменных в исходной регрессии.

Теоретический основой теста Бройша-Пагана является то, что если гетероскедастичность присутствует в данных, то квадраты остатков регрессии будут коррелированы с предикторами, которые вызывают эту гетероскедастичность. Поэтому регрессия квадратов остатков на все объясняющие переменные должна иметь значимый коэффициент детерминации.

Формулы для вычисления статистики теста Бройша-Пагана выглядят следующим образом:

$ H = nR^2 $

где n – количество наблюдений, а R^2 – коэффициент детерминации из регрессии квадратов остатков на все объясняющие переменные и константу.

Статистика теста H имеет распределение хи-квадрат с k – 1 степенями свободы, где k – количество объясняющих переменных в исходной модели.


Провёдем оценку остатков модели с параметрами p=4, q=5.

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

residuals = model2.resid
residuals = pd.Series(residuals)
print(residuals.describe())

fig, ax = plt.subplots(2, 1, figsize=(10,8))
plot_acf(residuals, lags=100, ax=ax[0])
plot_pacf(residuals, lags=100, ax=ax[1])
plt.show()
```

Можно заметить, что на графике автокорреляции дисперсия постепенно увеличивается. Нет значительных выбросов. Для того, чтобы убедится в гомоскедатичности остатков, проведём тест Бройша-Пагана при $\alpha$ = 0,05

```python
import statsmodels.api as sm
from statsmodels.stats.diagnostic import het_breuschpagan

X = points
X = sm.add_constant(X)

# Проводим тест Бройша-Пагана
bp_test = het_breuschpagan(residuals[:437], X[:437])

# Выводим результаты
print('F статистика: ', bp_test[0])
print('p-value: ', bp_test[1])
print('LM статистика: ', bp_test[2])
print('LM p-value: ', bp_test[3])
```

P-значение оказалось меньше $\alpha$. Гепотезу о гомоскедатичности остатков отвергаем

Дисперсия на графике автокорреляции увелчивается, но менее явно, чем на предыдущем. Явных выбросов нет. Похоже на гомоскедатичность. Проведем тест.
